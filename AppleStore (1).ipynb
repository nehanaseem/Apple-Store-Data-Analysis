```python
import csv
```

```python
# Load the dataset from a CSV file
open_file = open('AppleStore.csv')
read_file = csv.reader(open_file)
dataset = list(read_file)  # Convert the CSV data into a list of lists
```

# **Checking Data Types**

```python
# Check the index of the 'user_rating' column
user_rating_index = dataset[0].index('user_rating')
```

```python
# Print the type of user ratings to ensure they are correctly interpreted
for i in dataset[1:]:
    print(type(i[7]))
```

# **Convert User Ratings to Float**

```python
# Convert and print user ratings as float values
for i in dataset[1:]:
    print(float(i[7]))
```

# **Calculate Average User Rating**

```python
# Extract user ratings and calculate average
user_rating = [float(i[7]) for i in dataset[1:]]  # Convert ratings to float and store in list
SUM = sum(user_rating)  # Calculate the total sum of user ratings
AVG1 = SUM / len(user_rating)  # Calculate the average user rating
```

# **Calculate Average User Rating Version**

```python
# Extract user rating versions and calculate average
user_rating_ver = [float(i[8]) for i in dataset[1:]]  # Convert rating versions to float and store in list
SUM1 = sum(user_rating_ver)  # Calculate the total sum of rating versions
AVG2 = SUM1 / len(user_rating_ver)  # Calculate the average user rating version
```

# **User Ratings for Free vs Paid Apps**

```python
# Extract user ratings for free apps and calculate average
user_ratings = [float(i[7]) for i in dataset[1:] if float(i[4]) == 0]  # Filter and convert ratings for free apps
AVG3 = sum(user_ratings) / len(user_ratings)  # Calculate the average user rating for free apps
```

```python
# Extract user ratings for paid apps and calculate average
user_ratin = [float(i[7]) for i in dataset[1:] if float(i[4]) != 0]  # Filter and convert ratings for paid apps
AVG5 = sum(user_ratin) / len(user_ratin)  # Calculate the average user rating for paid apps
```

# **User Ratings for Specific Genres**

```python
# Extract user ratings for free apps in the 'Games' genre and calculate average
user_rater = [float(i[7]) for i in dataset[1:] if float(i[4]) == 0 and i[11] == 'Games']  # Filter and convert ratings for free games
AVG4 = sum(user_rater) / len(user_rater)  # Calculate the average user rating for free games
```

```python
# Extract user ratings for free apps in 'Games' or 'Music' genres, print details, and calculate average
user_ratig = [float(i[7]) for i in dataset[1:] if float(i[4]) == 0 and (i[11] == 'Games' or i[11] == 'Music')]  # Filter and convert ratings for free games and music
AVG6 = sum(user_ratig) / len(user_ratig)  # Calculate the average user rating for free games and music
```

# **Pandas DataFrame Exploration**

```python
import pandas as pd
```

```python
# Load dataset into a Pandas DataFrame
df = pd.read_csv('AppleStore.csv')
```

```python
# Display basic information about the DataFrame
df.info()  # Print summary of the DataFrame
df.describe()  # Print descriptive statistics of numerical columns
```

```python
# Display specific columns and rows from the DataFrame
print(df[['id', 'price', 'size_bytes']])  # Display selected columns
print(df.iloc[10:15])  # Display rows 10 to 14
print(df.iloc[10:15, 0:4])  # Display rows 10 to 14 and columns 0 to 3
print(df.iloc[[4, 7, 9], [0, 1, 7]])  # Display specific rows and columns
print(df.iloc[0:5, [0, 1, 7]])  # Display rows 0 to 4 and specific columns
print(df.loc[0])  # Display the first row
print(df.loc[4:7])  # Display rows 4 to 7
print(df.loc[4:7, 'currency'])  # Display rows 4 to 7 for the 'currency' column
```

# **Query Specific Conditions**

```python
# Query specific conditions
B_I = df['track_name'] == 'Bible'  # Filter rows where the track name is 'Bible'
print(df[B_I])  # Display rows where the track name is 'Bible'
print(df[df['prime_genre'] == 'Music']['price'])  # Display prices for apps in the 'Music' genre
print(df.loc[df['prime_genre'] == 'Music', ['price']])  # Display prices for apps in the 'Music' genre with labels
print(df['prime_genre'].value_counts())  # Count the number of apps in each genre
```
